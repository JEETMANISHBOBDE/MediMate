```python
import pandas as pd
import gradio as gr

# Load the dataset (ensure the CSV file is in the same directory as your script)
medicines_data = pd.read_csv("A_Z_medicines_dataset_of_India.csv")

# Expanded dictionary of symptoms and corresponding common OTC medicines
symptom_medicine_mapping = {
    "headache": {"medicine": "Paracetamol", "condition": "Tension headache or migraine"},
    "fever": {"medicine": "Paracetamol", "condition": "Flu or viral infection"},
    "cold": {"medicine": "Cetirizine", "condition": "Common cold or allergies"},
    "cough": {"medicine": "Dextromethorphan syrup", "condition": "Cough due to cold or throat irritation"},
    "stomach pain": {"medicine": "Pantoprazole", "condition": "Acidity or mild gastritis"},
    "diarrhea": {"medicine": "ORS + Loperamide", "condition": "Mild food poisoning or viral infection"},
    "vomiting": {"medicine": "Ondansetron", "condition": "Nausea or food poisoning"},
    "allergy": {"medicine": "Levocetirizine", "condition": "Seasonal allergies or mild allergic reaction"},
    "body pain": {"medicine": "Ibuprofen", "condition": "Muscle pain or mild inflammation"},
    "sore throat": {"medicine": "Strepsils lozenges or Betadine gargle", "condition": "Throat infection or irritation"},
    "indigestion": {"medicine": "Digene or Pantoprazole", "condition": "Acid reflux or overeating"},
    "constipation": {"medicine": "Isabgol (Psyllium Husk) or Dulcolax", "condition": "Irregular bowel movements"},
    "gas": {"medicine": "Simethicone", "condition": "Bloating or indigestion"},
    "acidity": {"medicine": "Ranitidine or Pantoprazole", "condition": "Acid reflux or heartburn"},
    "dizziness": {"medicine": "Electral (ORS)", "condition": "Dehydration or low blood pressure"},
    "weakness": {"medicine": "ORS or Glucose-D", "condition": "Dehydration or vitamin deficiency"},
    "skin rash": {"medicine": "Calamine lotion or Cetirizine", "condition": "Allergic reaction or skin irritation"},
    "eye irritation": {"medicine": "Refresh Tears (Lubricating Eye Drops)", "condition": "Dry eyes or mild allergy"},
    "ear pain": {"medicine": "Otex ear drops", "condition": "Ear infection or wax buildup"},
    "nasal congestion": {"medicine": "Otrivin Nasal Spray", "condition": "Blocked nose due to cold or sinusitis"},
    "back pain": {"medicine": "Diclofenac gel or Ibuprofen", "condition": "Muscle strain or posture-related pain"},
    "joint pain": {"medicine": "Volini spray or Ibuprofen", "condition": "Arthritis or mild inflammation"},
    "menstrual pain": {"medicine": "Meftal-Spas", "condition": "Menstrual cramps"},
    "sleeplessness": {"medicine": "Melatonin (OTC) or chamomile tea", "condition": "Insomnia or stress"},
    "stress": {"medicine": "Ashwagandha capsules", "condition": "Mild anxiety or stress"},
    "depression": {"medicine": "Consult a doctor for appropriate treatment", "condition": "Depression or mood disorder"},
    "sinusitis": {"medicine": "Saline nasal spray or Cetirizine", "condition": "Sinus infection or allergy"},
    "high blood pressure": {"medicine": "Amlodipine or Lisinopril", "condition": "Hypertension"},
    "heartburn": {"medicine": "Omeprazole or Ranitidine", "condition": "Gastroesophageal reflux disease (GERD)"},
    "acne": {"medicine": "Benzoyl Peroxide or Salicylic acid", "condition": "Acne or pimples"},
    "chronic fatigue": {"medicine": "Vitamin B12 supplements or Multivitamins", "condition": "Fatigue or vitamin deficiency"},
    "mild burns": {"medicine": "Burnol or Aloe vera gel", "condition": "Minor burns or skin irritation"},
    "allergic rhinitis": {"medicine": "Fluticasone nasal spray or Levocetirizine", "condition": "Seasonal allergies or allergic rhinitis"},
    "UTI (Urinary Tract Infection)": {"medicine": "Nitrofurantoin or Ciprofloxacin", "condition": "Urinary tract infection"},
    "itching": {"medicine": "Hydrocortisone cream or Calamine lotion", "condition": "Skin irritation or allergic reaction"},
    "swelling": {"medicine": "Diclofenac gel or Ibuprofen", "condition": "Injury, inflammation, or arthritis"},
    "high cholesterol": {"medicine": "Atorvastatin or Simvastatin", "condition": "Hypercholesterolemia"},
    "high blood sugar": {"medicine": "Metformin or Insulin", "condition": "Diabetes Mellitus"},
    "ringworm": {"medicine": "Clotrimazole or Miconazole", "condition": "Fungal skin infection"},
    "hemorrhoids": {"medicine": "Preparation H or Anusol", "condition": "Hemorrhoids or anal discomfort"},
    "pimples": {"medicine": "Benzoyl Peroxide or Salicylic acid", "condition": "Acne or oily skin"},
    "mouth ulcers": {"medicine": "Kenalog in Orabase or Bonjela", "condition": "Mouth ulcers or sores"},
    "nausea": {"medicine": "Dimenhydrinate or Ondansetron", "condition": "Motion sickness or food poisoning"},
    "bloating": {"medicine": "Activated charcoal or Simethicone", "condition": "Gas or indigestion"},
    "weight loss": {"medicine": "Consult a doctor for a proper diagnosis", "condition": "Unexplained weight loss"},
    "poor appetite": {"medicine": "Appetite stimulants or Multivitamins", "condition": "Loss of appetite or nutrient deficiency"},
    "asthma": {"medicine": "Salbutamol inhaler or Fluticasone inhaler", "condition": "Asthma"},
    "epilepsy": {"medicine": "Valproate or Lamotrigine", "condition": "Seizures or Epilepsy"},
    "stroke": {"medicine": "Thrombolytics, Antiplatelets, Anticoagulants", "condition": "Cerebrovascular accident (Stroke)"},
    "pneumonia": {"medicine": "Antibiotics like Amoxicillin or Azithromycin", "condition": "Pneumonia"},
    "chronic obstructive pulmonary disease (COPD)": {"medicine": "Bronchodilators, Steroids", "condition": "COPD"},
    "malaria": {"medicine": "Artemisinin-based combination therapy", "condition": "Malaria"},
    "hepatitis B": {"medicine": "Tenofovir or Entecavir", "condition": "Chronic Hepatitis B"},
    "hepatitis C": {"medicine": "Direct-acting antivirals like Sofosbuvir", "condition": "Chronic Hepatitis C"},
    "HIV/AIDS": {"medicine": "Antiretroviral therapy (ART)", "condition": "Human Immunodeficiency Virus"},
    "dengue": {"medicine": "Supportive care, Paracetamol", "condition": "Dengue fever"},
    "tuberculosis": {"medicine": "Isoniazid, Rifampicin", "condition": "Tuberculosis"},
    "lupus": {"medicine": "Hydroxychloroquine, Immunosuppressive drugs", "condition": "Systemic Lupus Erythematosus"},
    "thyroid disorder": {"medicine": "Levothyroxine or Antithyroid drugs", "condition": "Hypothyroidism or Hyperthyroidism"},
    "chickenpox": {"medicine": "Calamine lotion, Antihistamines", "condition": "Chickenpox"},
    "measles": {"medicine": "Vitamin A supplementation, Supportive care", "condition": "Measles"},
    "mumps": {"medicine": "Supportive care, Pain relievers", "condition": "Mumps"},
    "rubella": {"medicine": "Supportive care, Pain relievers", "condition": "Rubella (German Measles)"},
    "varicella": {"medicine": "Acyclovir or Supportive care", "condition": "Varicella (Chickenpox)"},
    "arthritis": {"medicine": "Methotrexate or Ibuprofen", "condition": "Rheumatoid arthritis or osteoarthritis"},
    "gout": {"medicine": "Colchicine or Allopurinol", "condition": "Gout"},
    "psoriasis": {"medicine": "Topical corticosteroids or Calcipotriene", "condition": "Psoriasis"},
    "eczema": {"medicine": "Hydrocortisone or Tacrolimus", "condition": "Eczema or atopic dermatitis"},
    "diabetes": {"medicine": "Metformin or Insulin", "condition": "Type 1 or Type 2 Diabetes"},
    "cancer": {"medicine": "Consult a doctor for chemotherapy or immunotherapy", "condition": "Various types of cancer"},
    "schizophrenia": {"medicine": "Antipsychotic medications like Olanzapine", "condition": "Schizophrenia"},
    "bipolar disorder": {"medicine": "Lithium or Valproate", "condition": "Bipolar disorder"},
    "Parkinson's disease": {"medicine": "Levodopa or Dopamine agonists", "condition": "Parkinson's disease"},
    "Alzheimer's disease": {"medicine": "Donepezil or Rivastigmine", "condition": "Alzheimer's disease"},
    "multiple sclerosis": {"medicine": "Interferons or Copaxone", "condition": "Multiple sclerosis"},
    "anemia": {"medicine": "Iron supplements or Vitamin B12", "condition": "Iron-deficiency or pernicious anemia"},
    "bronchitis": {"medicine": "Bronchodilators, Cough syrups", "condition": "Acute or chronic bronchitis"},
    "liver cirrhosis": {"medicine": "Liver transplant or Diuretics", "condition": "Cirrhosis of the liver"},
    "kidney stones": {"medicine": "Pain relief or Lithotripsy", "condition": "Kidney stones"},
    "osteoporosis": {"medicine": "Bisphosphonates or Calcium supplements", "condition": "Osteoporosis"}
}




# Enhanced function for a more interactive chatbot
import random
import datetime

# Tracking previous questions (to keep conversation context)
user_context = {}


# Enhanced function for a more interactive chatbot
def medicine_query(user_input, user_id="default_user"):
    user_input = user_input.strip().lower()

    # Get current time for dynamic greeting
    hour = datetime.datetime.now().hour
    if hour < 12:
        greeting = "Good morning! 🌞"
    elif 12 <= hour < 18:
        greeting = "Good afternoon! ☀️"
    else:
        greeting = "Good evening! 🌙"

    # Bot introduction
    if user_input in ["hi", "hello", "hey"]:
        user_context[user_id] = None  # Reset context
        return f"{greeting} I'm your medicine assistant. How can I help you today? 😊"

    # Handling general questions
    general_responses = {
        "how are you": "I'm just a bot, but I'm here to help you! How can I assist with your health concerns? 😊",
        "who are you": "I'm your friendly medicine assistant! I can suggest medicines based on symptoms or provide information about medicines. How can I help?",
        "what can you do": "I can suggest over-the-counter medicines based on symptoms and provide details about medicines. Just tell me how you're feeling!"
    }

    if user_input in general_responses:
        return general_responses[user_input]

    # Check if multiple symptoms are mentioned
    detected_symptoms = {symptom: info for symptom, info in symptom_medicine_mapping.items() if symptom in user_input}

    if detected_symptoms:
        follow_up_questions = [
            "How long have you been experiencing this symptom? 🤔",
            "Is it mild or severe? 🏥",
            "Have you tried any remedies yet? 🌿",
            "Would you like some general health tips as well? 🏋️"
        ]

        user_context[user_id] = "symptom_followup"  # Save context

        response = f" Based on what you've mentioned, here are my recommendations:\n\n"
        for symptom, info in detected_symptoms.items():
            response += f"- {info['condition']} (Symptom: {symptom})\n  Recommended Medicine: {info['medicine']}\n\n"

        response += "(Consult a doctor if symptoms persist or worsen.)\n\n"
        response += random.choice(follow_up_questions)  # Ask a follow-up question
        return response

    # Follow-up question responses
    if user_id in user_context:
        context = user_context[user_id]

        if context == "symptom_followup":
            # Responses based on follow-up answers
            if "days" in user_input or "week" in user_input or "long" in user_input:
                return "If it's been more than a few days, it might be best to consult a doctor. 🏥"
            elif "mild" in user_input:
                return "Since it's mild, rest and home remedies might help. Let me know if you need suggestions! 🌿"
            elif "severe" in user_input:
                return "Severe symptoms might need professional medical attention. Please consult a doctor ASAP. 🚑"
            elif "yes" in user_input:
                return "That's great! Natural remedies can be very effective. Let me know if you need medicine recommendations too. 😊"
            elif "no" in user_input:
                return "No worries! I can suggest some over-the-counter medicines. Let me know what you're experiencing. 💊"

            # Reset context after answering
            user_context[user_id] = None

    if user_input == "list":
        medicines_list = "\n".join(medicines_data['name'][:10])  # Sample 10 medicines
        return f"Here are some medicines in the dataset:\n{medicines_list}\n\n(Showing a sample. Ask about any specific medicine!)"

    # Search for the medicine in the dataset
    found = medicines_data[medicines_data['name'].str.lower().str.contains(user_input)]
    if not found.empty:
        response = []
        for _, row in found.iterrows():
            details = (
                f"Medicine: {row['name']}\n"
                f"Price: ₹{row['price(₹)']}\n"
                f"Manufacturer: {row['manufacturer_name']}\n"
                f"Type: {row['type']}\n"
                f"Packaging: {row['pack_size_label']}\n"
                f"Composition: {row['short_composition1']}\n"
            )
            if pd.notna(row['short_composition2']):
                details += f"Additional Composition: {row['short_composition2']}\n"
            details += f"Discontinued: {'Yes' if row['Is_discontinued'] else 'No'}"
            response.append(details)

        health_tips = [
            "💡 Tip: Always check the expiry date before taking any medicine!",
            "🌿 Staying hydrated helps medicines work better in your body.",
            "👌Practice Good Hygiene – Wash hands frequently to prevent infections.",
            "💪Regular Health Checkups – Detect and prevent health issues early.",
            "✔️Limit Sugar & Processed Foods – Opt for whole, natural foods."
        ]
        return "\n\n".join(response) + "\n\n" + random.choice(health_tips)

    return "Hmm... I couldn't find anything related to that. Can you try rephrasing your question? 🤔"


import gradio as gr

def chat_interface():
    with gr.Blocks() as chat_ui:
        gr.Markdown("<h1 style='font-size: 40px; text-align: center;'>🩺 MediMate</h1>")

        chatbot = gr.Chatbot(label="MediMate AI")  # Renamed the Chatbot tab
        user_input = gr.Textbox(placeholder="Type your message here...", show_label=False)
        send_button = gr.Button("Send")

        def respond(history, message):
            response = medicine_query(message)
            history.append((message, response))
            return history, ""

        send_button.click(respond, inputs=[chatbot, user_input], outputs=[chatbot, user_input])
        user_input.submit(respond, inputs=[chatbot, user_input], outputs=[chatbot, user_input])

    return chat_ui

chat_ui = chat_interface()
chat_ui.launch(share=True)


```
